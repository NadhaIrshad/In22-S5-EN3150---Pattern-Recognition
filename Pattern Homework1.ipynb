### 🔹 MinMaxScaler

MinMaxScaler scales all data features in the range **[0, 1]**, or **[-1, 1]** if negative values are present in the dataset. 

Use it when:
- The data does **not follow a normal distribution**
- You need scaled data for algorithms like **decision trees, k-nearest neighbors, or support vector machines**

> ⚠️ It works best when **outliers are minimal or absent**, as it is sensitive to extreme values.

---

### 🔧 **How it Works**

It scales data to a fixed range by subtracting the minimum value and dividing by the range (max - min), ensuring all feature values lie within the specified range.

\[
X_{\text{scaled}} = \frac{X - X_{\text{min}}}{X_{\text{max}} - X_{\text{min}}}
\]

Where:
- \( X \) = original feature value  
- \( X_{\text{min}} \) = minimum value of the feature  
- \( X_{\text{max}} \) = maximum value of the feature  

---

### ✅ Advantages
- Ensures all features have the **same scale**
- **Simple and easy to interpret**

### ❌ Disadvantages
- **Sensitive to outliers** – extreme values can distort scaling, causing most data points to cluster near 0 or 1
- **Fixed range** limits flexibility for datasets with changing scales

### 🔹 StandardScaler

**StandardScaler** is a feature scaling technique that transforms data to follow the **Standard Normal Distribution (SND)**.  
It standardizes numeric features by centering the data and scaling it to unit variance.

Use it when:
- Your model (like **SVM**, **Logistic Regression**, or **Neural Networks**) assumes the data is **normally distributed**.

---

### 🔧 **How it Works**

It subtracts the mean of the data and divides by the standard deviation, centering the data around 0 with a standard deviation of 1.

\[
X_{\text{scaled}} = \frac{X - \mu}{\sigma}
\]

Where:
- \( X \) = original value  
- \( \mu \) = mean of the feature  
- \( \sigma \) = standard deviation of the feature  

---

### ✅ Advantages
- Handles features with **different units** effectively  
- Reduces impact of outliers **without completely removing** them  

### ❌ Disadvantages
- **Sensitive to outliers** – extreme values can skew the mean and standard deviation  
- **Not ideal for non-normal distributions**


```python
from sklearn.preprocessing import RobustScaler

# Select two features
features = ['magnesium', 'malic_acid']
df_selected = df[features]

# Apply MinMaxScaler
scaler = RobustScaler()
scaled_data = scaler.fit_transform(df_selected)
df_scaled = pd.DataFrame(scaled_data, columns=[f"{col}_scaled" for col in features])

# Plot before and after scaling
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Original data histograms
axes[0, 0].hist(df_selected['magnesium'], bins=20)
axes[0, 0].set_title("Original Magnesium")
axes[0, 1].hist(df_selected['malic_acid'], bins=20)
axes[0, 1].set_title("Original Malic Acid")

# Scaled data histograms
axes[1, 0].hist(df_scaled['magnesium_scaled'], bins=20, color='green')
axes[1, 0].set_title("Robust Scaled Magnesium")
axes[1, 1].hist(df_scaled['malic_acid_scaled'], bins=20, color='green')
axes[1, 1].set_title("RObust Scaled Malic Acid")

plt.tight_layout()
plt.show()

# === PLOT SCATTER PLOTS ===
plt.figure(figsize=(12, 5))

# Original scatter
plt.subplot(1, 2, 1)
plt.scatter(df_selected['magnesium'], df_selected['malic_acid'], alpha=0.7)
plt.title("Original Data")
plt.xlabel("Magnesium")
plt.ylabel("Malic Acid")

# Scaled scatter
plt.subplot(1, 2, 2)
plt.scatter(df_scaled['magnesium_scaled'], df_scaled['malic_acid_scaled'], color='green', alpha=0.7)
plt.title("Robust Scaled Data")
plt.xlabel("Magnesium (Scaled)")
plt.ylabel("Malic Acid (Scaled)")

plt.tight_layout()
plt.show()

# === PLOT BOX PLOTS ===
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Original box plots
axes[0, 0].boxplot(df_selected['magnesium'], vert=True, patch_artist=True)
axes[0, 0].set_title("Original Magnesium")
axes[0, 1].boxplot(df_selected['malic_acid'], vert=True, patch_artist=True)
axes[0, 1].set_title("Original Malic Acid")

# Scaled box plots
axes[1, 0].boxplot(df_scaled['magnesium_scaled'], vert=True, patch_artist=True, boxprops=dict(color='green'))
axes[1, 0].set_title("Robust Scaled Magnesium")
axes[1, 1].boxplot(df_scaled['malic_acid_scaled'], vert=True, patch_artist=True, boxprops=dict(color='green'))
axes[1, 1].set_title("Robust Scaled Malic Acid")

plt.tight_layout()
plt.show()
```

### 💪 RobustScaler

RobustScaler reduces the impact of outliers by using the **median** and **interquartile range (IQR)** for scaling, instead of the mean and standard deviation. This makes it more robust to extreme values.

Use RobustScaler when:
- Your data contains **many outliers**
- You want to maintain the **relative distances** between non-outlier data points
- You're using algorithms that are **sensitive to outliers**, like K-Means, PCA, or linear models

---

### 📐 Formula:

\[
X_{\text{scaled}} = \frac{X - Q_1}{Q_3 - Q_1}
\]

Where:
- \(X\) = original feature value  
- \(Q_1\) = 25th percentile (first quartile)  
- \(Q_3\) = 75th percentile (third quartile)

---

### ✅ Advantages
- Resistant to outliers
- Maintains the structure of non-outlier data better than MinMaxScaler

### ❌ Disadvantages
- May perform poorly on **highly skewed data**
- Less interpretable than MinMaxScaler (doesn’t map to a fixed range like [0, 1])


#### Power Transformer 

Power transforms are a family of parametric, monotonic transformations that are applied to make data more Gaussian-like. This is useful for modeling issues related to heteroscedasticity (non-constant variance), or other situations where normality is desired.

Currently, PowerTransformer supports the Box-Cox transform and the Yeo-Johnson transform. The optimal parameter for stabilizing variance and minimizing skewness is estimated through maximum likelihood.

Box-Cox requires input data to be strictly positive, while Yeo-Johnson supports both positive or negative data.

By default, zero-mean, unit-variance normalization is applied to the transformed data.

```python
from sklearn.preprocessing import PowerTransformer

# Select two features
features = ['magnesium', 'malic_acid']
df_selected = df[features]

# Apply MinMaxScaler
scaler = PowerTransformer(method='yeo-johnson')
scaled_data = scaler.fit_transform(df_selected)
df_scaled = pd.DataFrame(scaled_data, columns=[f"{col}_scaled" for col in features])

# Plot before and after scaling
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Original data histograms
axes[0, 0].hist(df_selected['magnesium'], bins=20)
axes[0, 0].set_title("Original Magnesium")
axes[0, 1].hist(df_selected['malic_acid'], bins=20)
axes[0, 1].set_title("Original Malic Acid")

# Scaled data histograms
axes[1, 0].hist(df_scaled['magnesium_scaled'], bins=20, color='green')
axes[1, 0].set_title("Power Transoformed Magnesium")
axes[1, 1].hist(df_scaled['malic_acid_scaled'], bins=20, color='green')
axes[1, 1].set_title("Power Transformed Malic Acid")

plt.tight_layout()
plt.show()

# === PLOT SCATTER PLOTS ===
plt.figure(figsize=(12, 5))

# Original scatter
plt.subplot(1, 2, 1)
plt.scatter(df_selected['magnesium'], df_selected['malic_acid'], alpha=0.7)
plt.title("Original Data")
plt.xlabel("Magnesium")
plt.ylabel("Malic Acid")

# Scaled scatter
plt.subplot(1, 2, 2)
plt.scatter(df_scaled['magnesium_scaled'], df_scaled['malic_acid_scaled'], color='green', alpha=0.7)
plt.title("Power Transformed Data")
plt.xlabel("Magnesium (Scaled)")
plt.ylabel("Malic Acid (Scaled)")

plt.tight_layout()
plt.show()

# === PLOT BOX PLOTS ===
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Original box plots
axes[0, 0].boxplot(df_selected['magnesium'], vert=True, patch_artist=True)
axes[0, 0].set_title("Original Magnesium")
axes[0, 1].boxplot(df_selected['malic_acid'], vert=True, patch_artist=True)
axes[0, 1].set_title("Original Malic Acid")

# Scaled box plots
axes[1, 0].boxplot(df_scaled['magnesium_scaled'], vert=True, patch_artist=True, boxprops=dict(color='green'))
axes[1, 0].set_title("Power transformed Magnesium")
axes[1, 1].boxplot(df_scaled['malic_acid_scaled'], vert=True, patch_artist=True, boxprops=dict(color='green'))
axes[1, 1].set_title("Power Transformed Malic Acid")

plt.tight_layout()
plt.show()
```

### ⚡ PowerTransformer

PowerTransformer is a normalization technique that **makes data more Gaussian-like** (normal distribution). It applies a **power transformation** to each feature independently to **stabilize variance** and **minimize skewness**.

It's useful when:
- Your data is **highly skewed**
- You want to improve model performance for **linear models**, **PCA**, or **regression**
- You are working with **heteroscedastic data** (where variance changes with the mean)

---

### 🔧 Two Supported Methods:

1. **Yeo-Johnson** (works with both positive and negative values)  
2. **Box-Cox** (only for **positive** values)

---

### 📐 Formula (Box-Cox Transformation):

\[
X_{\text{scaled}} =
\begin{cases}
\frac{(X^\lambda - 1)}{\lambda}, & \text{if } \lambda \ne 0 \\
\ln(X), & \text{if } \lambda = 0
\end{cases}
\]

Where:
- \(X\) = original value (must be > 0 for Box-Cox)
- \(\lambda\) = transformation parameter automatically learned from data

### 📐 Formula (Yeo-Johnson Transformation):

\[
X_{\text{scaled}} =
\begin{cases}
\frac{[(X + 1)^\lambda - 1]}{\lambda}, & X \geq 0,\ \lambda \ne 0 \\
\ln(X + 1), & X \geq 0,\ \lambda = 0 \\
\frac{-[(−X + 1)^{2−\lambda} - 1]}{2 - \lambda}, & X < 0,\ \lambda \ne 2 \\
−\ln(−X + 1), & X < 0,\ \lambda = 2
\end{cases}
\]

---

### ✅ Advantages
- Reduces skewness and makes data more normal
- Helps many machine learning algorithms perform better

### ❌ Disadvantages
- More complex than other scalers
- May **not always improve** performance if data is already approximately normal
- Yeo-Johnson can distort data when unnecessary

### 📊 QuantileTransformer

The `QuantileTransformer` is a preprocessing method that transforms features to follow a **uniform** or **normal (Gaussian)** distribution. It uses the **quantile function (inverse CDF)** to map the original data to the target distribution.

---

#### ✅ How It Works

1. Computes the **empirical cumulative distribution function (CDF)** of each feature.
2. Maps the feature values to the corresponding quantiles.
3. Transforms these quantiles to a **target distribution**:
   - `'uniform'` (default)
   - `'normal'` (Gaussian)

---

#### 🔢 Transformation Formula

For a value \\( x \\):

\\[
x_{\text{scaled}} = F^{-1}(\hat{F}(x))
\\]

- \\( \hat{F}(x) \\) is the empirical CDF of the feature.
- \\( F^{-1} \\) is the quantile function (inverse CDF) of the desired output distribution.

---

#### 🎯 Use Cases

- When features are **non-Gaussian**.
- When data contains **outliers**.
- Preprocessing before **PCA** or models that assume normality.

---

#### ⚖️ Advantages

- Makes data **more Gaussian-like**.
- **Robust** to outliers (uses ranking).
- Helps improve performance of many machine learning models.

---

#### ⚠️ Disadvantages

- **Original distances** between samples are not preserved.
- Can **overfit** on small datasets if `n_quantiles` is too high.
